```python
import shutil, os, sys, pathlib, subprocess, time, re, torch
from google.colab import files

# ============================================================
# RESET — NO KERNEL SHUTDOWN (ini yang bikin crash loop)
# Cukup hapus folder manual, ga perlu do_shutdown
# ============================================================
for d in ['/content/Real-ESRGAN', '/content/ECCV2022-RIFE',
          '/content/results', '/content/frames',
          '/content/upscaled', '/content/interpolated']:
    shutil.rmtree(d, ignore_errors=True)

sys.path_importer_cache.clear()
print("💥 HyperGaruda V4.1 — Clean Start (No Kernel Crash)")

# ======================== KONFIGURASI ========================
REPO_ESRGAN  = pathlib.Path("/content/Real-ESRGAN")
REPO_RIFE    = pathlib.Path("/content/ECCV2022-RIFE")
WEIGHTS_DIR  = REPO_ESRGAN / "weights"
FRAMES_DIR   = pathlib.Path("/content/frames")
UPSCALED_DIR = pathlib.Path("/content/upscaled")
INTERP_DIR   = pathlib.Path("/content/interpolated")
RESULTS_DIR  = pathlib.Path("/content/results")

MODEL_NAME     = "realesr-animevideov3"
MODEL_FILE     = WEIGHTS_DIR / f"{MODEL_NAME}.pth"
MODEL_URL      = "https://github.com/xinntao/Real-ESRGAN/releases/download/v0.2.5.0/realesr-animevideov3.pth"
# Updated to Google Drive file ID for flownet.pkl
RIFE_MODEL_GD_ID = "1h42aGYPNJn2q8j_GVkS_yDu__G_UZ2GX"

UPSCALE    = 4
SUFFIX     = "HyperGaruda_x4_60fps"
FFMPEG_BIN = "ffmpeg"
TARGET_FPS = 60   # Ganti 120 kalau mau 120fps

# ========== DETEKSI HARDWARE ==========
cpu_cores     = os.cpu_count() or 2
gpu_available = torch.cuda.is_available()
gpu_mem_total = torch.cuda.get_device_properties(0).total_memory / 1e9 if gpu_available else 0
gpu_name      = torch.cuda.get_device_name(0) if gpu_available else "None"

# CUDA Optimization
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.benchmark        = True
torch.backends.cudnn.allow_tf32       = True

# TILE — safety net biar ga OOM crash
# T4 15GB + 1080x1920 frame = bisa OOM kalau TILE=0
# TILE=512 = balanced, ga crash, tetep fast
TILE = 512

print(f"🔍 Hardware  : CPU={cpu_cores}c | {gpu_name} ({gpu_mem_total:.1f}GB)")
print(f"⚙️  Config    : TILE={TILE} | FP16 | TARGET_FPS={TARGET_FPS}")
print(f"🚀 CUDA      : cuDNN benchmark=True | TF32=True")

# ============================================================
# HELPERS
# ============================================================
def run_cmd(cmd, cwd=None, check=True, capture=False):
    return subprocess.run(cmd, cwd=cwd, check=check,
                          capture_output=capture, text=capture)

def print_section(title):
    print(f"\n{'='*78}\n{title}\n{'='*78}")

def set_cpu_env():
    for var in ["OMP_NUM_THREADS","MKL_NUM_THREADS","OPENBLAS_NUM_THREADS",
                "VECLIB_MAXIMUM_THREADS","NUMEXPR_NUM_THREADS"]:
        os.environ[var] = str(cpu_cores)
    print(f"🧵 CPU saturated: {cpu_cores} threads")

def get_video_fps(path):
    r = run_cmd([FFMPEG_BIN, "-i", str(path)], capture=True, check=False)
    m = re.search(r'(\d+(?:\.\d+)?)\s*fps', r.stderr)
    if m: return float(m.group(1))
    m2 = re.search(r'(\d+(?:\.\d+)?)\s*tbr', r.stderr)
    return float(m2.group(1)) if m2 else 30.0

def get_video_res(path):
    r = run_cmd([FFMPEG_BIN, "-i", str(path)], capture=True, check=False)
    m = re.search(r'(\d{3,4})x(\d{3,4})', r.stderr)
    return (int(m.group(1)), int(m.group(2))) if m else (None, None)

# ============================================================
# STEP 1: UPLOAD
# ============================================================
def upload_video():
    print_section("STEP 1: UPLOAD VIDEO")
    uploaded   = files.upload()
    name       = next(iter(uploaded.keys()))
    input_path = pathlib.Path("/content") / name

    for d in [RESULTS_DIR, FRAMES_DIR, UPSCALED_DIR, INTERP_DIR]:
        d.mkdir(parents=True, exist_ok=True)

    w, h = get_video_res(input_path)
    fps  = get_video_fps(input_path)
    size = input_path.stat().st_size / 1e6

    print(f"✅ Input  : {name}")
    print(f"📹 Info   : {w}x{h} @ {fps:.0f}fps | {size:.1f}MB")
    print(f"🎯 Output : {(w or 0)*UPSCALE}x{(h or 0)*UPSCALE} @ {TARGET_FPS}fps")
    return input_path, fps

# ============================================================
# STEP 2: SYSTEM DEPS
# ============================================================
def install_system_deps():
    print_section("STEP 2: SYSTEM DEPENDENCIES")
    run_cmd(["apt-get", "update", "-qq"])
    run_cmd(["apt-get", "install", "-y", "-qq", "ffmpeg"])
    print("✅ FFmpeg ready")

# ============================================================
# STEP 3: CLONE
# ============================================================
def clone_repos():
    print_section("STEP 3: CLONE REPOS")
    if not REPO_ESRGAN.exists():
        run_cmd(["git", "clone", "-q", "https://github.com/xinntao/Real-ESRGAN.git"])
        print("✅ Real-ESRGAN cloned")
    else:
        print("✅ Real-ESRGAN ada")

    if not REPO_RIFE.exists():
        run_cmd(["git", "clone", "-q", "https://github.com/hzwer/ECCV2022-RIFE.git"])
        print("✅ RIFE cloned")
    else:
        print("✅ RIFE ada")

# ============================================================
# STEP 4: INSTALL DEPS — NO requirements.txt RIFE
# ============================================================
def install_deps():
    print_section("STEP 4: INSTALL DEPENDENCIES")

    print("🔄 PyTorch CUDA...")
    run_cmd([sys.executable, "-m", "pip", "install", "-q", "--upgrade", "pip"])
    run_cmd([sys.executable, "-m", "pip", "install", "-q",
             "torch", "torchvision", "torchaudio",
             "--index-url", "https://download.pytorch.org/whl/cu118"])

    print("🔄 Real-ESRGAN deps...")
    run_cmd([sys.executable, "-m", "pip", "install", "-q",
             "-r", "requirements.txt"], cwd=str(REPO_ESRGAN))
    run_cmd([sys.executable, "-m", "pip", "install", "-q",
             "basicsr", "facexlib", "gfpgan"])
    run_cmd([sys.executable, "-m", "pip", "install", "-q", "-e", "."],
            cwd=str(REPO_ESRGAN))

    # RIFE — SKIP requirements.txt (numpy pinned lama = crash)
    # Install manual aja yang dibutuhkan
    print("🔄 RIFE deps (manual, skip requirements.txt)...")
    run_cmd([sys.executable, "-m", "pip", "install", "-q",
             "numpy",
             "opencv-python-headless",
             "scikit-image",
             "tqdm",
             "imageio",
             "imageio-ffmpeg"])
    print("✅ All deps installed")

    # Install gdown for Google Drive downloads
    print("🔄 Installing gdown...")
    run_cmd([sys.executable, "-m", "pip", "install", "-q", "gdown"])
    print("✅ gdown installed")

    # CUDA check
    r = run_cmd([sys.executable, "-c",
        "import torch; "
        "print(f'CUDA: {torch.cuda.is_available()} | "
        "GPU: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else None} | "
        "PyTorch: {torch.__version__}')"],
        capture=True)
    print(r.stdout.strip())

# ============================================================
# STEP 4b: FIX BASICSR
# ============================================================
def fix_basicsr():
    print_section("STEP 4b: FIX BASICSR")
    import glob
    for p in glob.glob("/usr/local/lib/python*/dist-packages/basicsr/data/degradations.py"):
        try:
            c = open(p).read()
            old = "from torchvision.transforms.functional_tensor import rgb_to_grayscale"
            new = "from torchvision.transforms.functional import rgb_to_grayscale"
            if old in c:
                open(p, 'w').write(c.replace(old, new))
                print(f"✅ Patched: {p}")
            else:
                print("✅ No patch needed")
        except Exception as e:
            print(f"⚠️  {e}")

    with open("/content/dynamic_fix.py", "w") as f:
        f.write("""
import sys, types
from torchvision.transforms.functional import rgb_to_grayscale
ft = types.ModuleType('torchvision.transforms.functional_tensor')
ft.rgb_to_grayscale = rgb_to_grayscale
sys.modules['torchvision.transforms.functional_tensor'] = ft
print("✅ Dynamic fix applied")
""")
    subprocess.run([sys.executable, "/content/dynamic_fix.py"])

# ============================================================
# STEP 5: DOWNLOAD MODELS
# ============================================================
def download_models():
    print_section("STEP 5: DOWNLOAD MODELS")

    WEIGHTS_DIR.mkdir(parents=True, exist_ok=True)
    if not MODEL_FILE.exists():
        run_cmd(["wget", "-q", MODEL_URL, "-O", str(MODEL_FILE)])
        print("✅ Real-ESRGAN model downloaded")
    else:
        print("✅ Real-ESRGAN model ada")

    rife_dir = REPO_RIFE / "train_log"
    rife_dir.mkdir(parents=True, exist_ok=True)
    rife_pkl = rife_dir / "flownet.pkl"
    if not rife_pkl.exists():
        print("🔄 RIFE flownet.pkl (Google Drive)...")
        # Use gdown for Google Drive download
        run_cmd(["gdown", "--id", RIFE_MODEL_GD_ID, "-O", str(rife_pkl)])
        print("✅ RIFE model downloaded")
    else:
        print("✅ RIFE model ada")

# ============================================================
# STEP 6a: EXTRACT FRAMES
# ============================================================
def extract_frames(input_path):
    print_section("STEP 6a: EXTRACT FRAMES")
    set_cpu_env()

    cmd = [
        FFMPEG_BIN, "-i", str(input_path),
        "-threads", str(cpu_cores),
        "-q:v", "1",
        "-vsync", "0",
        f"{FRAMES_DIR}/%08d.png"   # NO -pix_fmt = FFmpeg auto
    ]
    print("🔄 Extracting frames...")
    subprocess.run(cmd, check=True)
    count = len(list(FRAMES_DIR.glob("*.png")))
    print(f"✅ {count} frames extracted")
    return count

# ============================================================
# STEP 6b: REAL-ESRGAN UPSCALE
# ============================================================
def run_upscale():
    print_section(f"STEP 6b: REAL-ESRGAN UPSCALE (TILE={TILE})")

    script = REPO_ESRGAN / "inference_realesrgan.py"
    if not script.exists():
        raise FileNotFoundError(f"Script tidak ada: {script}")

    cmd = [
        sys.executable, str(script),
        "-i", str(FRAMES_DIR),
        "-o", str(UPSCALED_DIR),
        "-n", MODEL_NAME,
        "-s", str(UPSCALE),
        "--suffix", "",
        "-t", str(TILE),   # 512 = safe, no OOM
    ]

    print(f"🚀 GPU upscaling — TILE={TILE} | FP16 | {UPSCALE}x")
    result = subprocess.run(cmd, cwd=str(REPO_ESRGAN))
    if result.returncode != 0:
        raise RuntimeError(f"ESRGAN gagal: exit {result.returncode}")

    count = len(list(UPSCALED_DIR.glob("*.png")))
    print(f"✅ {count} frames upscaled {UPSCALE}x")

# ============================================================
# STEP 6c: RIFE INTERPOLATION
# ============================================================
def run_rife(orig_fps):
    import math
    print_section(f"STEP 6c: RIFE ({orig_fps:.0f}fps → {TARGET_FPS}fps)")

    multiplier = TARGET_FPS / orig_fps
    passes     = max(1, math.ceil(math.log2(max(multiplier, 1.01))))
    actual_fps = orig_fps * (2 ** passes)

    print(f"📊 {orig_fps:.0f} → {actual_fps:.0f}fps ({passes} pass, x{2**passes})")

    # Cari script RIFE
    candidates = [
        REPO_RIFE / "inference_video.py",
        REPO_RIFE / "inference_img.py",
        *sorted(REPO_RIFE.glob("*inference*.py"))
    ]
    script = next((s for s in candidates if s.exists()), None)

    if not script:
        print("⚠️  RIFE script tidak ditemukan — skip")
        return str(UPSCALED_DIR), orig_fps

    cmd = [
        sys.executable, str(script),
        "--img", str(UPSCALED_DIR),
        "--output", str(INTERP_DIR),
        "--multi", str(int(2 ** passes)),
        "--fp16",
    ]

    print(f"🎞️  RIFE running {passes} pass...")
    result = subprocess.run(cmd, cwd=str(REPO_RIFE))

    if result.returncode != 0:
        print("⚠️  RIFE gagal — fallback ke upscaled frames")
        return str(UPSCALED_DIR), orig_fps

    count = len(list(INTERP_DIR.glob("*.png")))
    print(f"✅ {count} frames → {actual_fps:.0f}fps")
    return str(INTERP_DIR), actual_fps

# ============================================================
# STEP 6d: ENCODE
# ============================================================
def encode_video(input_path, frames_dir, out_fps):
    print_section(f"STEP 6d: ENCODE @ {out_fps:.0f}fps")

    final = RESULTS_DIR / f"{input_path.stem}_{SUFFIX}.mp4"
    cmd = [
        FFMPEG_BIN,
        "-framerate", str(out_fps),
        "-i", f"{frames_dir}/%08d.png",
        "-i", str(input_path),
        "-map", "0:v", "-map", "1:a?",
        "-c:v", "libx264",
        "-preset", "fast",
        "-crf", "18",
        "-pix_fmt", "yuv420p",
        "-c:a", "copy",
        "-threads", str(cpu_cores),
        str(final), "-y"
    ]

    print(f"🔄 Encoding → {final.name}...")
    subprocess.run(cmd, check=True)
    size = final.stat().st_size / 1e6
    print(f"✅ {final.name} ({size:.1f}MB)")
    return final

# ============================================================
# MAIN
# ============================================================
def main():
    print("\n" + "🔥" * 39)
    print("  HYPERGARUDA V4.1 — POLYPHONY T4")
    print("  Real-ESRGAN 4x + RIFE Interpolation")
    print(f"  No kernel crash. No OOM. Pure W.")
    print("🔥" * 39)

    t0 = time.time()

    input_path, orig_fps = upload_video()
    install_system_deps()
    clone_repos()
    install_deps()
    fix_basicsr()
    download_models()

    frame_count          = extract_frames(input_path)
    run_upscale()
    frames_dir, out_fps  = run_rife(orig_fps)
    final                = encode_video(input_path, frames_dir, out_fps)

    elapsed = time.time() - t0

    print_section("STEP 7: DOWNLOAD")
    print(f"⏱️  Time    : {elapsed:.0f}s ({elapsed/60:.1f} min)")
    print(f"🎞️  Frames  : {frame_count} @ {orig_fps:.0f}fps → {out_fps:.0f}fps")
    print(f"📐 Scale   : {UPSCALE}x")
    print(f"🎮 HW      : {gpu_name} | TILE={TILE} | FP16")

    files.download(str(final))

    print("\n" + "🏁" * 39)
    print("  FINISH LINE — HyperGaruda V4.1")
    print("🏁" * 39)

if __name__ == "__main__":
    main()
```
